```python

import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['figure.figsize'] = [20, 10]    # resize figure
 
def isNaN(num):
    return num!= num
```

```python
# Define constants
g = 9.8
 
# Define initial values
time = 0.0
dt = 0.000001  # time step in seconds
i = 1  # index for looping
maxTime = 100.0  # maximum time in seconds
mass = 1.0  # kg
# kParam = 5.0e-4  # drag parameter equivalent to Cd*rho*area/2
angle = 89.9  # degrees
v = 30.0  # initial speed in m/s
x = 0.0  # initial x position in meters
y = 0.0  # initial y position in meters
xarray = []
yarray = []
timearray = []
```

```python
Cd = input('Enter drag coefficient: ')
area = input('Enter cross-sectional area: ')
rho = input('Enter density: ')
kParam = float(Cd)*float(rho)*float(area)/2.0
```

```python
# Break initial speed into components
angle = np.radians(angle)
vx = v * np.cos(angle)
vy = v * np.sin(angle)
 
# Begin iterating while y > 0
while y >= 0 and time < maxTime:
  x = x + vx * dt
  y = y + vy * dt
  angle = np.arctan(vy/vx)
 
  vx = vx - ( kParam/mass ) * ( vx*vx + vy*vy ) * np.cos(angle) * dt
  vy = vy - ( kParam/mass ) * ( vx*vx + vy*vy ) * np.sin(angle) * dt - g * dt
  
  i += 1
  if i % 10 == 0:
    timearray.append(time)
    xarray.append(x)
    yarray.append(y)
 
  time = time + dt
 
# plot y versus x position
plt.plot(xarray,yarray)
plt.show
```
